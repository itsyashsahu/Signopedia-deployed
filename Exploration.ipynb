```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import os

#Loading Training Data

data=[]
labels=[]
num_classes=43
cur_dir=os.getcwd()

for i in range(num_classes):
    path=os.path.join(cur_dir,"data","Train",str(i))
    images=os.listdir(path)
    for a in images:
        try:
            im_path=os.path.join(path,a)
            image=Image.open(im_path)
            image=image.resize((30,30))
            image=np.array(image)
            data.append(image)
            labels.append(i)
        except:
            print("Could not open image")

data=np.array(data)
labels=np.array(labels)

print("Done Loading Images!!!")
```

```python
#Loading Test Data

test=[]
cur_dir=os.getcwd()

path=os.path.join(cur_dir,"data","Test")
images=os.listdir(path)
for a in images:
    try:
        im_path=os.path.join(path,a)
        image=Image.open(im_path)
        image=image.resize((30,30))
        image=np.array(image)
        test.append(image)
    except:
        print("Could not open image")

test=np.array(test)

print("Done Loading Test Images!!!")
```
